```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_excel('/Users/kspann/Desktop/NeoBank/Oxford/250709 OxfordColostrum_Fadil_REPORT.xlsx')
df
```

```python
# extract numeric day from 'PP day' and create 'unique_sample_id'
day_num = df['PP day'].astype(str).str.extract(r'(\d+)', expand=False)
df['unique_sample_id'] = df['Participant ID'].fillna('NA').astype(str) + '_day' + day_num.fillna('NA')
```

```python
df.columns
```

```python
df[['Sample Name#', 'Participant ID', 'PP day', 'unique_sample_id']]
```

```python
# remove leading/trailing whitespace from Participant ID
df['Participant ID'] = df['Participant ID'].str.strip()
```

```python
# count unique participant IDs
unique_non_null = df['Participant ID'].nunique(dropna=True)   # excludes NaN
unique_including_na = df['Participant ID'].nunique(dropna=False)  # counts NaN as a distinct value

print("Unique Participant IDs (excluding NaN):", unique_non_null)
print("Unique Participant IDs (including NaN):", unique_including_na)
```

```python
df['unique_sample_id'].nunique()
```

```python
# find duplicated unique_sample_id values and show counts and corresponding rows
dup_counts = df['unique_sample_id'].value_counts()
duplicates = dup_counts[dup_counts > 1]

print("Duplicate unique_sample_id values and their counts:")
print(duplicates)

# show all rows that have a duplicated unique_sample_id
if not duplicates.empty:
    dup_rows = df[df['unique_sample_id'].isin(duplicates.index)].sort_values('unique_sample_id')
    display(dup_rows[['Sample Name#', 'Participant ID', 'PP day', 'unique_sample_id']])
else:
    print("No duplicates found in unique_sample_id.")
```
